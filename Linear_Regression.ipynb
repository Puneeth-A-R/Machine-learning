```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], marker='+', c='green')
plt.show()
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
reg = linear_model.LinearRegression()
reg.fit(new_df.values, df['per capita income (US$)'].values)
```

```python
res = (reg.predict([[2022]]))
print(res)
print(f"m={reg.coef_}, c={reg.intercept_}")
```
